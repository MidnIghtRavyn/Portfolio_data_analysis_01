## Distribusi Biaya Asuransi (Insurance Charges)

Visualisasi histogram digunakan untuk melihat sebaran nilai biaya asuransi pada dataset.

Insight yang diperoleh:

- Sebagian besar individu memiliki biaya asuransi pada rentang rendah hingga menengah.
- Distribusi data terlihat condong ke kanan (right skewed), yang menunjukkan adanya sejumlah kecil individu dengan biaya sangat tinggi.
- Hal ini mengindikasikan bahwa meskipun mayoritas pelanggan memiliki biaya normal, terdapat kelompok dengan risiko kesehatan tinggi yang menyebabkan biaya meningkat secara signifikan.

Distribusi yang tidak simetris ini umum terjadi pada data biaya kesehatan karena adanya faktor risiko tertentu seperti usia, kebiasaan merokok, dan kondisi kesehatan.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['charges'], kde=True)

plt.title("Distribution of Insurance Charges")
plt.show()
```

## Analisis Hubungan Usia dengan Biaya Asuransi

Visualisasi scatter plot digunakan untuk melihat hubungan antara usia (age) dan biaya asuransi (charges).

Insight yang diperoleh:

- Terdapat kecenderungan bahwa semakin bertambah usia, biaya asuransi meningkat.
- Pola ini menunjukkan adanya korelasi positif antara usia dan biaya.
- Namun terlihat beberapa kelompok biaya yang berbeda, yang kemungkinan dipengaruhi faktor lain seperti status perokok.

Hal ini menunjukkan bahwa usia merupakan salah satu faktor yang mempengaruhi biaya asuransi kesehatan.

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='age', y='charges', data=df)

plt.title("Age vs Charges")
plt.show()
```

## Analisis Status Perokok terhadap Biaya Asuransi

Boxplot digunakan untuk membandingkan distribusi biaya asuransi antara perokok dan non-perokok.

Insight yang diperoleh:

- Individu yang merokok memiliki biaya asuransi jauh lebih tinggi dibandingkan non-perokok.
- Variasi biaya pada kelompok perokok juga lebih besar.
- Terdapat banyak outlier pada kelompok non-perokok, namun tetap lebih rendah dibanding perokok.

Hal ini menunjukkan bahwa status merokok merupakan faktor risiko utama yang mempengaruhi biaya asuransi kesehatan.

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='smoker', y='charges', data=df)

plt.title("Smoker vs Charges")
plt.show()
```

## Analisis BMI terhadap Biaya Asuransi berdasarkan Status Perokok

Scatter plot dengan kategori warna berdasarkan status perokok digunakan untuk melihat interaksi antara BMI dan biaya asuransi.

Insight yang diperoleh:

- Pada kelompok perokok, kenaikan BMI cenderung diikuti peningkatan biaya asuransi secara signifikan.
- Pada kelompok non-perokok, hubungan BMI dan biaya tidak terlalu kuat.
- Perokok dengan BMI tinggi memiliki biaya paling mahal dibanding kelompok lainnya.

Hal ini menunjukkan adanya kombinasi risiko antara BMI tinggi dan kebiasaan merokok terhadap biaya kesehatan.

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)

plt.title("BMI vs Charges by Smoker Status")
plt.show()
```

## Analisis Korelasi Antar Variabel (Correlation Matrix)

Correlation matrix digunakan untuk melihat hubungan antar variabel numerik dalam dataset serta mengetahui faktor mana yang paling berpengaruh terhadap biaya asuransi (charges).

Insight yang diperoleh:

- Variabel **smoker_flag** memiliki korelasi paling tinggi terhadap biaya asuransi, yang menunjukkan bahwa status merokok merupakan faktor utama yang mempengaruhi besarnya biaya kesehatan.
- Variabel **age** dan **BMI** juga memiliki korelasi positif terhadap biaya, meskipun tidak sekuat status merokok.
- Variabel lain seperti jumlah anak memiliki pengaruh yang relatif kecil terhadap biaya asuransi.

Analisis korelasi ini memperkuat temuan pada visualisasi sebelumnya bahwa faktor gaya hidup dan risiko kesehatan memiliki peran penting dalam menentukan biaya asuransi.

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```
